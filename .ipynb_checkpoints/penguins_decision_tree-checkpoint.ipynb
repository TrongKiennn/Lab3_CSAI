### Loading Libraries and Data

```python

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import graphviz
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from IPython.display import display
```

### Load and preprocess the penguins dataset

```python
penguins = sns.load_dataset("penguins").dropna()
features = penguins[['bill_length_mm', 'flipper_length_mm', 'body_mass_g', 'sex']]
labels = penguins['species']
features = pd.get_dummies(features, columns=['sex'])
penguins_data = features.copy()
penguins_data['species'] = labels
penguins_data = penguins_data.sample(frac=1, random_state=42).reset_index(drop=True)
```

### Splitting the Dataset into Multiple Train/Test Ratios

```python
def split_data(data, test_size):
    X = data.drop(columns=['species'])
    y = data['species']
    return train_test_split(X, y, test_size=test_size, random_state=42, stratify=y)

ratios = [0.4, 0.6, 0.8, 0.9]
datasets = {}
for ratio in ratios:
    X_train, X_test, y_train, y_test = split_data(penguins_data, test_size=1 - ratio)
    datasets[f"{round(ratio*100)}/{round((1-ratio)*100)}"] = {
        'X_train': X_train,
        'X_test': X_test,
        'y_train': y_train,
        'y_test': y_test
    }
```

### Visualizing Class Distributions

```python
for key, data in datasets.items():
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))
    data['y_train'].value_counts().plot(kind='bar', ax=ax[0], title=f'{key} - Train')
    data['y_test'].value_counts().plot(kind='bar', ax=ax[1], title=f'{key} - Test')
    plt.suptitle(f'Class Distribution ({key})')
    plt.show()
```

### Building and Visualizing Decision Trees

```python
for key, data in datasets.items():
    clf = DecisionTreeClassifier(criterion='entropy', random_state=42)
    clf.fit(data['X_train'], data['y_train'])

    dot_data = export_graphviz(
        clf,
        out_file=None,
        feature_names=data['X_train'].columns,
        class_names=[str(cls) for cls in clf.classes_],
        filled=True,
        rounded=True,
        special_characters=True
    )

    graph = graphviz.Source(dot_data)
    print(f"Decision Tree ({key})")
    display(graph)

```

### Evaluating the Decision Tree Classifiers

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

for key, data in datasets.items():
    clf = DecisionTreeClassifier(criterion='entropy', random_state=42)
    clf.fit(data['X_train'], data['y_train'])
    y_pred = clf.predict(data['X_test'])

    print(f"Dataset {key}")
    print("Classification Report:\n", classification_report(data['y_test'], y_pred))
    print("Confusion Matrix:\n", confusion_matrix(data['y_test'], y_pred))
    print("="*50)

    cm = confusion_matrix(data['y_test'], y_pred)
    class_names = clf.classes_
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix Heatmap ({key} split)')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
```

### Analyzing the Effect of Tree Depth

```python
depth_results = []
depths = [None, 2, 3, 4, 5, 6, 7]
X_train = datasets['80/20']['X_train']
y_train = datasets['80/20']['y_train']
X_test = datasets['80/20']['X_test']
y_test = datasets['80/20']['y_test']

for d in depths:
    clf = DecisionTreeClassifier(criterion='entropy', max_depth=d, random_state=42)
    clf.fit(X_train, y_train)
    acc = accuracy_score(y_test, clf.predict(X_test))
    depth_results.append((d, acc))

    # Visualization with Graphviz
    dot_data = export_graphviz(
        clf,
        out_file=None,
        feature_names=X_train.columns,
        class_names=clf.classes_,
        filled=True,
        rounded=True,
        special_characters=True
    )
    graph = graphviz.Source(dot_data)
    print(f"📌 Decision Tree (max_depth={d})")
    display(graph)

depth_df = pd.DataFrame(depth_results, columns=["max_depth", "Accuracy"])
print(depth_df)
plt.plot(depth_df["max_depth"].astype(str), depth_df["Accuracy"], marker="o")
plt.title("Accuracy vs Max Depth")
plt.xlabel("Max Depth")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()
```

### Insights on Tree Depth and Accuracy

The experiment conducted with different values of the `max_depth` parameter (None, 2, 3, 4, 5, 6, 7) on the 80/20 training and test split reveals the following insights:

1. **Accuracy varies slightly with tree depth.**  
   When the tree is restricted to a small depth, such as `max_depth = 2`, it achieves the lowest accuracy (0.955). This suggests underfitting, where the tree is too shallow to capture the patterns in the data.

2. **The highest accuracy (0.985) is achieved at multiple depths.**  
   Specifically, `max_depth = `4`, `5`, `6`, and `7` all result in the same maximum accuracy. This indicates that once the tree reaches a sufficient depth (≥ 4), adding more depth does not improve performance on the test set.

3. **Moderate depth already performs well.**  
   Even `max_depth = 3` gives relatively high accuracy (0.970), showing that a tree of modest complexity can still generalize well.

4. **Overall trend: accuracy stabilizes after depth = 4.**  
   The model reaches peak performance at `max_depth = 4`, and increasing the depth beyond that yields no gain. This flat plateau suggests the decision boundaries are not very complex beyond that point.
